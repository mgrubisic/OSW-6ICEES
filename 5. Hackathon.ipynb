## <font color=blue>5.2. What is our challenge?</font>

As a newbie in OpenSees(py), the best learning methodology is practicing. You will practice a nice example of OpenSees modeling and Python programming. Please consider this hackathon not as challange but funny game to play. There is not any loser at the end.

In this hackathon, please, complete the tasks given below;

**Task 1.** Develop an elastic model (name as model_100) of the bridge pier in OpenSeesPy in Figure 1a and estimate the first natural period, compare with your hand-calculation for validation [1]. 

**Task 2.** On the elastic model of the bridge pier (model_100), conduct time-history analysis [2] in order to determine the elastic seismic demand at the plastic hinge section [3].

**Task 3.** Develop en inelastic model of the bridge pier [4]. Assign the yield point value as found in **Task 2**.

Remembering the definition of the ductility ($\mu$), one can obtain the displacement ductility of the system, Figure 1b. Now considering the variation of the yield point, one can also change the ductility ($\mu$) ratio. 

**Task 4.** Develop a small loop in your OpenSeesPy model, developed in **Task 3**, where yield point is iterated in order to obtain the displacement ductility $\mu = 2 \pm 0.1$. 

Notes: 

[1] Weight of the bridge deck is 30% of the axial load capacity of the pier $( W = (F_c \times A_g) \times 0.30 )$. Take $(g = 9.81 m/s^2)$.

[2] Use the ground motion record of RSN126_GAZLI_GAZ090_SF_1.316.txt, Figure 2, in hackathon.

[3] Plastic hinge length is the half of the cross-section dimension $( L_P = \phi/2 )$. 

[4] Use Material2 as explained in 3.4. Uniaxial Material Testing via Static Analysis.

## <font color=blue>5.4. What are the Key Performance Indicators (KPI)?</font>

In this hackathon there are various Key Performance Indicators (KPI) as listed below: 

1. The most important KPI is yoursel. If you feel that you are on right track learning OpenSees(py), you got the first KPI. Kudos to you. 

2. The first minor KPI is the comparison of the natural vibration periods obtained from hand-calculation and OpenSees model. 

3. The second minor is the elastic seismic demand at the critical section. 

4. The first major KPI is the calculation of the displacement ductility of the bridge pier. 

5. The second major KPE is the iteration of the yield point for target ductility. This KPI will reveal your understanding not only OpenSees but also programming progress.
